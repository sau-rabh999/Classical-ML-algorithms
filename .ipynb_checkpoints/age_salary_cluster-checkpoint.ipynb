```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('age_salary_cluster.csv')
df
```

```python
scaler = MinMaxScaler()
scaler.fit(df[['Age']])
scaler.fit(df[['EstimatedSalary']])
df['Age'] = scaler.transform(df[['Age']])
df['EstimatedSalary'] = scaler.transform(df[['EstimatedSalary']])
df
```

```python
plt.scatter(scaler.inverse_transform([df.Age]), scaler.inverse_transform([df.EstimatedSalary]))
```

```python
#finding no. of clusters using elbow method

k_rng = range(1,10)
sse = []
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['Age', 'EstimatedSalary']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('Sum of Squared Error')
plt.plot(k_rng, sse)
```

```python
km = KMeans(n_clusters=4)
km
```

```python
y_pred = km.fit_predict(df[['Age', 'EstimatedSalary']])
y_pred
```

```python
df['Cluster'] = y_pred
df
```

```python
km.cluster_centers_
```

```python
df1 = df[df.Cluster==0]
df2 = df[df.Cluster==1]
df3 = df[df.Cluster==2]
df4 = df[df.Cluster==3]

plt.scatter(scaler.inverse_transform([df1.Age]), scaler.inverse_transform([df1.EstimatedSalary]), color='red', label='income lvl 1')
plt.scatter(scaler.inverse_transform([df2.Age]), scaler.inverse_transform([df2.EstimatedSalary]), color='green', label='income lvl 2')
plt.scatter(scaler.inverse_transform([df3.Age]), scaler.inverse_transform([df3.EstimatedSalary]), color='blue', label='income lvl 3')
plt.scatter(scaler.inverse_transform([df4.Age]), scaler.inverse_transform([df4.EstimatedSalary]), color='yellow', label='income lvl 4')
plt.scatter(scaler.inverse_transform([km.cluster_centers_[:,0]]),scaler.inverse_transform([km.cluster_centers_[:,1]]), color='purple', marker='*', label='centroid')

plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
```
